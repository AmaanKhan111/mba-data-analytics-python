# 03 - How to rename columns?


```python
df = pd.DataFrame({"A Coloumn" : [1,2,3,4,5,6,7,8,],"B Coloumn": [8,7,6,5,4,3,2,1,]})
df
```

```python
df.rename(columns={"A Coloumn": "1 Coloumn", "B Coloumn": "2 Coloumn",},inplace = True)
df
```

```python
# rename coloumns
df.columns=["A Columns","B Columns"]
df
```

```python
import pandas as pd
df = pd.DataFrame({"A Coloumn" : [1,2,3,4,5,6,7,8,],"B Coloumn": [8,7,6,5,4,3,2,1,]})
df
```

```python
# To replace any character , string
df.columns=df.columns.str.replace(" ","_")
df
```

```python
# Adding prefix to colums 
df = df.add_prefix("AK")
df
```

```python
df = df.add_suffix("AK")
df
```

```python
df.columns=["A_Column","B_Column"]
df
```
